```python
total_deaths = covid_df.new_deaths.sum() #sum of new_deaths column
total_cases = covid_df.new_cases.sum() #sum of new_cases column
total_tests = covid_df.new_tests.sum() #sum of new_tests column

print("Total Deaths: {:.0f}\nTotal Cases: {:.0f}\nTotal Tests: {:.0f}".format(total_deaths, total_cases, total_tests))
```

#### Extract data from datetime

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year #extracting the year from the date column
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month #extracting the month from the date column
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day #extracting the day from the date column
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday #extracting the weekday from the date column

# Convert Int to Weekday
weekday_map = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
covid_df['weekday'] = covid_df['weekday'].map(weekday_map)
covid_df.head()
```

```python
may_covid = covid_df[covid_df.month == 5] #filtering the rows with month 5
may_covid.head()
```

```python
may_covid_metrics = may_covid[['new_cases', 'new_deaths', 'new_tests']] #selecting the columns new_cases, new_deaths, new_tests
total_may_cases, total_may_deaths, total_may_tests = may_covid_metrics.sum() #sum of new_cases, new_deaths, new_tests
print("Total Cases in May: {:.0f}\nTotal Deaths in May: {:.0f}\nTotal Tests in May: {:.0f}".format(total_may_cases, total_may_deaths, total_may_tests))
```

```python
# Average new cases for Sunday
may_sunday_mean_cases = may_covid[may_covid.weekday == 'Sunday'].new_cases.mean()
print("Average new cases for Sundays in May: {:.2f}".format(may_sunday_mean_cases))
```
