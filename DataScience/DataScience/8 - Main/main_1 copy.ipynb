#### Extract data from datetime

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year #extracting the year from the date column
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month #extracting the month from the date column
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day #extracting the day from the date column
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday #extracting the weekday from the date column

# Convert Int to Weekday
weekday_map = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
covid_df['weekday'] = covid_df['weekday'].map(weekday_map)
covid_df.head()
```

```python
may_covid = covid_df[covid_df.month == 5] #filtering the rows with month 5
may_covid.head()
```

```python
may_covid_metrics = may_covid[['new_cases', 'new_deaths', 'new_tests']] #selecting the columns new_cases, new_deaths, new_tests
total_may_cases, total_may_deaths, total_may_tests = may_covid_metrics.sum() #sum of new_cases, new_deaths, new_tests
print("Total Cases in May: {:.0f}\nTotal Deaths in May: {:.0f}\nTotal Tests in May: {:.0f}".format(total_may_cases, total_may_deaths, total_may_tests))
```

```python
# Average new cases for Sunday
may_sunday_mean_cases = may_covid[may_covid.weekday == 'Sunday'].new_cases.mean()
print("Average new cases for Sundays in May: {:.2f}".format(may_sunday_mean_cases))
```

#### Grouping and aggregation

```python
covid_month_df = covid_df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum() #grouping the dataframe by month and summing the new_cases, new_deaths, new_tests
covid_month_df
```

```python
covid_weekday_mean_df = covid_df.groupby('weekday')[['new_cases', 'new_deaths', 'new_tests']].mean() #grouping the dataframe by month and finding the mean of new_cases, new_deaths, new_tests
covid_weekday_mean_df
```

#### Aggregate with cumulative sum

Acumulate all the rows in the same day

```python
covid_df['total_cases'] = covid_df.new_cases.cumsum() #cumulative sum of new_cases
covid_df['total_deaths'] = covid_df.new_deaths.cumsum() #cumulative sum of new_deaths
covid_df['total_tests'] = covid_df.new_tests.cumsum() #cumulative sum of new_tests
covid_df[200:210]
```

#### Merge multiple sources

```python
localtion_df = pd.read_csv('Datasets/locations.csv') #reading the locations.csv file
localtion_df.head()
```

```python
covid_df['location'] = 'Italy' #adding a new column location with value Italy
covid_df.head()
```

```python
merged_df = covid_df.merge(localtion_df, on='location') #merging the covid_df and location_df on location column
print(merged_df.shape)
print(covid_df.shape)
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population #calculating the cases per million
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population #calculating the deaths per million
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population #calculating the tests per million
merged_df.cases_per_million[200:210]
```

```python
result_df = merged_df[['date',
                          'new_cases', 
                          'total_cases', 
                          'new_deaths', 
                          'total_deaths', 
                          'new_tests', 
                          'total_tests', 
                          'cases_per_million', 
                          'deaths_per_million', 
                          'tests_per_million'
                        ]]
result_df.head()
```

```python
result_df.to_csv('Datasets/results.csv', index=None) #saving the result_df to results.csv file, pandas will add index column by default
```
